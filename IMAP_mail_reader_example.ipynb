```python
# 개념 1
# IMAP -> "메일을 읽고" 도와주는 규칙 
# 메일 접속 -> 메일 보내고 -> 메일 읽고 

# 개념 2 
# library 설치 -> plugin 설치 
```

```python
# 라이브러리 설치 단계 
import imaplib
import email
from email.header import decode_header

# 계정 설정 단계 (이메일 주소, 이메일 비번)
email_id = "pythontesthaeun@gmail.com"
password = "bzgailackktgibfe" # 2단계 인증 후 생성한 앱 비밀번호

# IMAP 설정, 로그인 단계
imap = imaplib.IMAP4_SSL("imap.gmail.com")
imap.login(email_id, password)

# 이메일 함 읽기
imap.select("INBOX")
_, messages = imap.search(None, "ALL")
# _, messages = imap.uid('search', None, '(FROM "pythontesthaeun@gmail.com")')
messages = messages[0].split() # [b'2', b'7', b'8']

# 반복문 돌며 메일 읽기 
for i in messages:
    _, data = imap.fetch(i, "(RFC822)")
    message = email.message_from_bytes(data[0][1])
    
    # 메일 제목 encoding 
    subject, encoding = decode_header(message["Subject"])[0]
    if encoding == 'utf-8':
        subject = subject.decode(encoding)
    
    print("제목: ", subject)
    print("날짜: ", message.get('Date'))
    print("발신인: ", message.get('From'))
    print("수신인: ", message.get('To'))
    
    # 메일 내용을 탐색하면서, 본문이 text 인 경우에만, 출력해라.
    for part in message.walk(): 
        if part.get_content_type() == 'text/plain':
            print("본문: ", part.get_payload(decode=True).decode())
    
    print("============")
    
imap.close()
```
